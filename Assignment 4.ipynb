```python
# Step 1: Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
# Step 2: Generate a dataset with normal distribution parameters
mean = 100
std_dev = 4
dataset_size = 100000
```

```python
# Generate the dataset
data = np.random.normal(loc=mean, scale=std_dev, size=dataset_size)
```

```python
# Step 3: Plotting the histogram and PDF
plt.figure(figsize=(12, 6))
sns.histplot(data, bins=50, kde=False, stat='density', color='skyblue', alpha=0.6)
xmin, xmax = plt.xlim()
```

```python
# Overlay the PDF
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'r', linewidth=2)
```

```python
plt.title('Normal Distribution (Mean = 100, Std Dev = 4)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid()
plt.show()
```
